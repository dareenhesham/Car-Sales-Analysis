##Outliers Detection


```python
numeric_cols = data.select_dtypes(include='number').columns
colors = '#B73A3A'
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=data, y=col, color=colors)
    plt.title(f'Box Plot for {col}')
    plt.tight_layout()
    plt.show()
```

##Correlation analysis


```python
x = data.copy()
label_encoder = LabelEncoder()
for col in x.columns:
    if x[col].dtype == 'object':
       x[col] = label_encoder.fit_transform(x[col])
correlation_matrix = x.corr()
```

```python
gray_gradient_colors = ['#F4F4F4', '#D1D1D1', '#888888', '#666666', '#474646']
gray_gradient_cmap = LinearSegmentedColormap.from_list("gray_gradient", gray_gradient_colors)

plt.figure(figsize=(15, 10), facecolor='none')
ax = sns.heatmap(correlation_matrix, annot=True, cmap=gray_gradient_cmap, fmt=".2f",linewidths=0.4, linecolor='white', square=True, cbar_kws={'shrink': 0.8})
ax.set_facecolor("none")

plt.title("Correlation Matrix", fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

#Data Cleaning

```python
data = data.dropna(subset=['Customer Name']).copy()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data = data.drop(columns=['Car_id', 'Phone'])
```

#Feature engineering

```python
def income_category(income):
    if income < 30000:
        return 'Low'
    elif income < 70000:
        return 'Medium'
    else:
        return 'High'

data['Income_Category'] = data['Annual Income'].apply(income_category)
```

```python
data['Year'] = data['Date'].dt.year
print(data[['Date', 'Year']].head())
```

```python
print(data.info())
```

```python
cars_by_model = data.groupby('Model').size().reset_index(name='Number of Cars Sold')
print(cars_by_model)
```

#Saving Data

```python
with pd.ExcelWriter('car_analysis.xlsx') as writer:
    data.to_excel(writer, sheet_name='car_analysis', index=False)
    cars_by_model.to_excel(writer, sheet_name='Cars_by_Model', index=False)
```
